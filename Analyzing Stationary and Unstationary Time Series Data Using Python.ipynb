**Analyzing Stationary and Unstationary Time Series Data Using Python**

>To understand and demonstrate the concept of stationarity in time series data by comparing the properties of generated stationary and non-stationary time series.

This Involves:
1. **Generating:** Creating synthetic datasets that represent both
stationary and non-stationary time series.
2.**Analyzing:** Calculating and visualizing key statistics (mean, variance) and the Autocorrelation Function (ACF) for both types of data.
3.**Illustrating:** Using rolling window calculations to show how mean and variance behave differently over time for stationary versus non-stationary data.
4. **Visualizing:** Creating plots to visually highlight the characteristics of each data type and the results of the analysis.

**Generating Stationary Data**


* A random dataset is generated using np.random.normal. Stationary data has constant mean and variance over time.
* The code calculates and prints the mean and variance of this data.
* It plots the data to visualize its behavior.
* It calculates and plots the ACF to see the correlation of the data with lagged versions of itself. For stationary data, the ACF should quickly drop to zero.
* It calculates and plots a rolling mean and rolling variance to further illustrate the consistent mean and variance over time

```python
# Let's Generate Random Data

data = np.random.normal(loc=0, scale=1, size=1000)
print(data)
```

```python
# Mean of the Random Data
mean= np.mean(data)
print("Mean:",mean)
```

```python
# Variance of the Data
variance = np.var(data)
print("Variance:", variance)
```

```python
import matplotlib.pyplot as plt
```

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
```

>**statsmodels.graphics.tsaplots.plot_acf**: Used to plot the Autocorrelation Function (ACF).

>**statsmodels.tsa.stattools.acf**: Used to calculate the ACF values.

```python
np.random.seed(0)
data= np.random.normal(loc=0,scale=1,size=100)
plt.plot(data)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Random Stationary Data')
plt.show()

```

Output


```python
print('variance of data:', np.var(data))
```

```python
acf_data = acf(data, fft=False)
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(data, ax=ax, lags=20)
plt.show()
```

```python
rolling_mean = np.convolve(data, np.ones(20)/20, mode='valid')


plt.plot(data, label='Data')
plt.plot(rolling_mean, label='Rolling Mean')
plt.title('Random Stationary Data Mean')
plt.legend()
plt.show()
```

```python
df= pd.DataFrame(data)
variance =df.rolling(window=20).var()
plt.plot(variance)
plt.title('Variance of Random Data on a 20 Moving Window')
plt.xlabel('Observation')
plt.ylabel('Variance')
plt.title('Random Stationary Data Variance')
plt.show()
print('variance of data:', np.var(data))
```

```python
acf_data = acf(data, fft=False)
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(data, ax=ax, lags=20)
plt.show()
```

**NON-STATIONARY DATA**

* A non-stationary dataset is generated by adding a linear trend (2 * t) to random noise.
* Non-stationary data has characteristics (like mean or variance) that change over time.
* The code plots this non-stationary data, showing the clear trend.
* It calculates and plots the rolling mean and rolling variance, demonstrating how these characteristics change over time for non-stationary data.
* It calculates and plots the ACF. For non-stationary data, the ACF typically decays slowly.

```python
# Generate non-stationary data
np.random.seed(0)
n = 100
t = np.arange(n)
y = 2 * t + np.random.normal(0, 1, n)
```

```python
# Plot the time series
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Random Non-Stationary Data')
plt.show()
```

```python
# Plot data and rolling mean

rolling_mean = np.convolve(y, np.ones(20)/20, mode='valid')

plt.plot(data, label='Data')
plt.plot(rolling_mean, label='Rolling Mean')
plt.title('Random non-Stationary Data Mean')
plt.legend()
plt.show()
```

```python
# Print mean of data
print('Mean of data:', np.mean(data))
```

```python
# Calculate the variance on a rolling window of 20 observations
variance = df.rolling(window=20).var()
```

```python
# Print variance of data
print('variance of data:', np.var(data))
```

```python
# Plot the variance
plt.plot(variance)
plt.title('Variance of Random Data on a 20 Moving Window')
plt.xlabel('Observation')
plt.ylabel('Variance')
plt.title('Random non-Stationary Data Variance')
plt.show()
```

```python
import statsmodels.api as sm
```

**statsmodels.api (as sm):**
>Used for time series analysis, specifically for calculating the ACF.

```python
# Compute ACF and plot
acf = sm.tsa.stattools.acf(y, nlags=n-1)
plt.stem(acf)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('ACF Plot of Random Non-Stationary Data')
plt.show()
```
